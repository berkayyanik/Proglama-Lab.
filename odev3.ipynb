```python
from sympy import Symbol,pprint
import sympy as sym
import sympy.plotting as syp
import matplotlib.pyplot as plt
```

```python
x = Symbol('x')
lamda = Symbol('lambda') #lambda olayın kaç kez tekrar ettiğini temsil eder
```

```python
my_f_part1 = sym.exp(-lamda)
pprint(my_f_part1)
```

```python
my_f_part2 = lamda**x
pprint(my_f_part2)
```

```python
pay = my_f_part1 * my_f_part2
pprint(pay)
```

```python
my_f_part3 = sym.factorial(x)
payda = my_f_part3
pprint(payda)
```

```python
sonuc = pay/payda
pprint(sonuc) #Poisson Distribution Formula
```

```python
sym.plot(sonuc.subs({lamda:5}),(x,0,10),title = 'Sympy Poisson Distribution Plot')
```

```python
#Matplotlib Poisson Distribution Plot
x1=[]
y1=[]
for deger in range(0,11):
    y=sonuc.subs({lamda:5,x:deger}).evalf()
    y1.append(y)
    x1.append(deger)
plt.plot(x1,y1)
```
